## 1.1 O que é Cabotagem?

Cabotagem é o transporte marítimo de cargas realizado entre portos de um mesmo país. No Brasil, essa modalidade é muito utilizada devido à extensa costa, sendo uma alternativa ao transporte rodoviário. A cabotagem é mais sustentável e econômica para grandes volumes de carga.

## 1.2 O que é TEU?

TEU (Twenty-foot Equivalent Unit) é uma unidade de medida padrão no transporte marítimo, usada para descrever a capacidade de um contêiner ou o volume de carga transportado. Um TEU equivale ao tamanho de um contêiner de 20 pés.

Primeiramente, importamos todas as bibliotecas que usaremos no projeto, instalamos o Prophet e criamos uma **SparkSession**, que é o ponto de entrada para usar a API do Spark

Realizamos o carregamento de dados históricos de **carga** e **atracação** em formato `.txt` para posterior análise de dados de cabotagem. Usamos o **PySpark** para processar e integrar os dados em um DataFrame.

Realizamos o Merge entre as duas tabelas, Carga e Atracacao, pelo IDAtracacao, e por fim filtramos apenas os dados com Tipo de Operação igual a Cabotagem e removemos algumas colunas que não usaremos nas análises.

```python
# Importações para PySpark
from pyspark.sql import SparkSession
from pyspark.sql.functions import (
    col, to_timestamp, unix_timestamp, year, month, regexp_replace, isnull, to_date, date_format, sum as _sum
)
from pyspark.sql import functions as F

# Importações para visualização e análise de dados 
import matplotlib.pyplot as plt
import pandas as pd

# Modelagem com Prophet 
from prophet import Prophet

# Logging para depuração 
import logging

spark = SparkSession.builder.appName("PrevisaoDemandaCarga").getOrCreate()

carga_file_paths = [
    "/FileStore/tables/2019Carga.txt",
    "/FileStore/tables/2020Carga.txt",
    "/FileStore/tables/2021Carga.txt",
    "/FileStore/tables/2022Carga.txt",
    "/FileStore/tables/2023Carga.txt",
    "/FileStore/tables/2024Carga.txt"
]

atracacao_file_paths = [
    "/FileStore/tables/2019Atracacao.txt",
    "/FileStore/tables/2020Atracacao.txt",
    "/FileStore/tables/2021Atracacao.txt",
    "/FileStore/tables/2022Atracacao.txt",
    "/FileStore/tables/2023Atracacao.txt",
    "/FileStore/tables/2024Atracacao.txt"
]

# Função para carregar e concatenar arquivos usando Spark
def carregar_e_concatenar_spark(file_paths):
    df = spark.read.csv(file_paths, sep=';', header=True, encoding='UTF-8')
    return df

# Carregar os dados de carga
df_carga = carregar_e_concatenar_spark(carga_file_paths)

# Filtrar somente os dados de Cabotagem
df_carga = df_carga.filter(df_carga['Tipo Operação da Carga'] == 'Cabotagem')

# Carregar os dados de atracação
df_atracacao = carregar_e_concatenar_spark(atracacao_file_paths)

# Realizar o merge
df_cabotagem = df_carga.join(df_atracacao, on='IDAtracacao', how='inner')

# Remover as colunas especificadas
colunas_excluir = [
    #'Data Chegada', 'Data Desatracação', 'Data Início Operação', 'Data Término Operação', 
    'Nº da Capitania', 'Instalação Portuária em Rio', 'SGUF', 'Município', 
    'Tipo de Navegação da Atracação', 'FlagMCOperacaoAtracacao', 'Nº do IMO', 'Ano', 'Mes'
]

df_cabotagem = df_cabotagem.drop(*colunas_excluir)

# Verificar o esquema do DataFrame resultante
df_cabotagem.printSchema()
```

Obtemos algumas informações do nosso data frame de cabotagem, podendo ver que temos o equivalente a mais de *17 GB* de dados.

```python
# Tamanho dos dados em bytes
def calcular_tamanho(df):
    return df.rdd.map(lambda row: len(str(row))).sum()

# Número de linhas e colunas
num_linhas = df_cabotagem.count()
num_colunas = len(df_cabotagem.columns)

# Tamanho em bytes
tamanho_em_bytes = calcular_tamanho(df_cabotagem)

# Exibir as informações
print("Informações do DataFrame df_cabotagem:")
print(f"- Tamanho (peso) dos dados: {tamanho_em_bytes} bytes")
print(f"- Número de linhas: {num_linhas}")
print(f"- Número de colunas: {num_colunas}")
```

Começamos removendo as linhas duplicadas, tratando valores nulos e formatando corretamente as principais colunas que serão utilizadas no modelo de previsão.

```python
# Removendo linhas duplicadas
df_cabotagem = df_cabotagem.dropDuplicates()

# Convertendo 'Data Atracação' para datetime, tratando erros
# Especificando o formato correto da data
df_cabotagem = df_cabotagem.withColumn(
    'Data Atracação', to_timestamp('Data Atracação', 'dd/MM/yyyy HH:mm:ss')
)

# Removendo dados de final de 2018 presentes nos arquivos de 2019
df_cabotagem = df_cabotagem.filter(year(col('Data Atracação')) >= 2019)

# Convertendo 'TEU' para numérico, tratando erros
# Substituindo vírgulas por pontos, se necessário, e convertendo para double
df_cabotagem = df_cabotagem.withColumn(
    'TEU', regexp_replace('TEU', ',', '.').cast('double')
)

# Removendo linhas com valores NaN após a conversão
df_cabotagem = df_cabotagem.na.drop(subset=['TEU', 'Data Atracação'])
```

# 6. Início de desenvolvimento dos KPIs



Agrupamos os dados por tipo de contêiner e somamos a quantidade de TEUs para cada grupo, ordenamos os resultados em ordem decrescente, selecionando os 10 primeiros. Criamos um gráfico de barras para visualizar de forma clara quais são os tipos de contêiner mais utilizados.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Agrupando por 'Tipo Conteiner' e somando 'TEU'
df_top10 = (df_mercadoria_cabotagem
            .groupBy("Tipo Conteiner")
            .agg(_sum("TEU").alias("TotalTEU"))
            .orderBy(col("TotalTEU").desc())
            .limit(10))  # Selecionando os 10 maiores

# Convertendo para Pandas para visualização
df_top10_pd = df_top10.toPandas() 

# Verificar o tipo dos dados de TotalTEU para garantir que seja numérico
df_top10_pd['TotalTEU'] = pd.to_numeric(df_top10_pd['TotalTEU'], errors='coerce')

# Configurando estilo do gráfico
sns.set(style="whitegrid")

# Criando o gráfico de barras horizontais
plt.figure(figsize=(10, 6))

# Plotando o gráfico
sns.barplot(
    x="TotalTEU",  # Quantidade de TEU no eixo x
    y="Tipo Conteiner",  # Tipos de Contêiner no eixo y
    data=df_top10_pd, 
    palette="viridis"
)

# Ajustando os detalhes do gráfico
plt.title("Top 10 Tipos de Contêiner por Quantidade de TEU", fontsize=16)
plt.xlabel("Quantidade de TEU", fontsize=14)
plt.ylabel("Tipo de Contêiner", fontsize=14)

# Ajustando os limites do eixo X manualmente, caso necessário
plt.xlim(0, df_top10_pd['TotalTEU'].max() * 1.1)  # Garantir que o eixo x esteja correto

# Ajustando o layout para evitar sobreposição
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```

Nós convertemos colunas de datas para o formato de timestamp e filtrando registros entre 2019 e 2024. Calculamos indicadores-chave como o tempo total no porto, tempo de operação, tempo de espera, tempo pós-operação e eficiência operacional. Em seguida, convertemos os dados para um DataFrame Pandas, criamos períodos mensais e calculamos a média do tempo total no porto para cada mês. 

```python

df_cabotagem = df_cabotagem.withColumn(
    'Data Chegada', to_timestamp('Data Chegada', 'dd/MM/yyyy HH:mm:ss')
).withColumn(
    'Data Desatracação', to_timestamp('Data Desatracação', 'dd/MM/yyyy HH:mm:ss')
).withColumn(
    'Data Início Operação', to_timestamp('Data Início Operação', 'dd/MM/yyyy HH:mm:ss')
).withColumn(
    'Data Término Operação', to_timestamp('Data Término Operação', 'dd/MM/yyyy HH:mm:ss')
)

# Filtrar dados de 2019 a 2024
df_cabotagem = df_cabotagem.filter((year(col('Data Chegada')) >= 2019) & (year(col('Data Chegada')) <= 2024))

# Recalcular os KPIs
df_cabotagem = df_cabotagem.withColumn(
    'Tempo_Total_Porto',
    (unix_timestamp('Data Desatracação') - unix_timestamp('Data Chegada')) / 3600
)

df_cabotagem = df_cabotagem.withColumn(
    'Tempo_Operacao',
    (unix_timestamp('Data Término Operação') - unix_timestamp('Data Início Operação')) / 3600
)

df_cabotagem = df_cabotagem.withColumn(
    'Tempo_Espera',
    (unix_timestamp('Data Início Operação') - unix_timestamp('Data Chegada')) / 3600
)

df_cabotagem = df_cabotagem.withColumn(
    'Tempo_Pos_Operacao',
    (unix_timestamp('Data Desatracação') - unix_timestamp('Data Término Operação')) / 3600
)

df_cabotagem = df_cabotagem.withColumn(
    'Eficiencia_Operacional',
    (col('Tempo_Operacao') / col('Tempo_Total_Porto')) * 100
)

# Filtrando e convertendo para Pandas
df_pandas = df_cabotagem.select('Data Chegada', 'Tempo_Total_Porto').toPandas()
df_pandas['Data Chegada'] = pd.to_datetime(df_pandas['Data Chegada'])

df_pandas['Mes_Ano'] = df_pandas['Data Chegada'].dt.to_period('M')
media_tempo_porto = df_pandas.groupby('Mes_Ano')['Tempo_Total_Porto'].mean().reset_index()

# Verificando o DataFrame filtrado
print(media_tempo_porto)

```

A partir do gráfico evolutido de Tempo Médio Total no Porto notamos uma alta média de espera principalmente no começo de 2019 e final de 2020.

```python
import matplotlib.pyplot as plt

# Ajustar o tamanho da figura
plt.figure(figsize=(15,7))

# Personalizar a linha do gráfico
plt.plot(media_tempo_porto['Mes_Ano'].astype(str), media_tempo_porto['Tempo_Total_Porto'], 
         marker='o', color='#1f77b4', markersize=8, linewidth=2, label='Tempo Total Porto')

# Adicionar título e rótulos com fontes maiores e em negrito
plt.title('Tempo Médio Total no Porto por Mês', fontsize=16, fontweight='bold', color='#333333')
plt.xlabel('Mês/Ano', fontsize=14, fontweight='bold', color='#333333')
plt.ylabel('Tempo (Horas)', fontsize=14, fontweight='bold', color='#333333')

# Personalizar os ticks do eixo X e Y
plt.xticks(rotation=45, fontsize=12, color='#555555')
plt.yticks(fontsize=12, color='#555555')

# Adicionar uma grade mais suave e estilizada
plt.grid(True, linestyle='--', alpha=0.7, color='#b3b3b3')

# Ajustar o layout para evitar sobreposição
plt.tight_layout()

# Adicionar uma legenda
plt.legend(loc='upper left', fontsize=12)

# Exibir o gráfico
plt.show()
```

Calculamos o tempo média de espera por porto e plotamos no gráfico, podendo concluir que o porto de Itaqui segue disparado como o porto de maior tempo médio de espera.

```python
# Calcular tempo médio de espera por porto
df_porto_espera = df_cabotagem.groupBy('Porto Atracação').agg(
    F.count('IDAtracacao').alias('Movimentacao'),
    F.mean('Tempo_Espera').alias('Tempo_Medio_Espera')
).orderBy('Movimentacao', ascending=False)

# Selecionar os top 10 portos com maior movimentação
top_portos = df_porto_espera.limit(10).toPandas()

# Ordenar pelo tempo médio de espera
top_portos = top_portos.sort_values(by='Tempo_Medio_Espera', ascending=False)

plt.figure(figsize=(12,8))
plt.barh(top_portos['Porto Atracação'], top_portos['Tempo_Medio_Espera'], color='skyblue')
plt.xlabel('Tempo Médio de Espera (Horas)')
plt.ylabel('Porto de Atracação')
plt.title('Top 10 Portos com Maior Tempo Médio de Espera')
plt.gca().invert_yaxis()  # Inverter o eixo y para que o maior valor fique no topo
plt.tight_layout()
plt.show()

```

Calculamos as médias mensais de Tempo_Espera e Tempo_Operacao e exibe um gráfico de barras comparativo. Ele agrupa os dados por mês/ano e visualiza as tendências ao longo do tempo, destacando variações nos tempos médios de espera e operação dos navios.

```python
df_pandas = df_cabotagem.select('Data Chegada', 'Tempo_Espera', 'Tempo_Operacao').toPandas()
df_pandas['Data Chegada'] = pd.to_datetime(df_pandas['Data Chegada'])
df_pandas['Mes_Ano'] = df_pandas['Data Chegada'].dt.to_period('M')

media_tempos = df_pandas.groupby('Mes_Ano')[['Tempo_Espera', 'Tempo_Operacao']].mean().reset_index()

media_tempos.plot(x='Mes_Ano', y=['Tempo_Espera', 'Tempo_Operacao'], kind='bar', figsize=(15,7))
plt.title('Tempo Médio de Espera vs. Operação por Mês')
plt.xlabel('Mês/Ano')
plt.ylabel('Tempo (Horas)')
plt.xticks(rotation=45)
plt.legend(['Tempo de Espera', 'Tempo de Operação'])
plt.tight_layout()
plt.show()
```

Criamos um histograma para visualizar a distribuição da eficiência operacional dos navios. Ele plota as frequências em 50 intervalos (bins) e destaca a média da eficiência com uma linha vertical vermelha. O gráfico ajuda a identificar a concentração dos valores e possíveis variações na eficiência operacional.

```python
df_pandas = df_cabotagem.select('Eficiencia_Operacional').toPandas()

plt.figure(figsize=(12,6))
plt.hist(df_pandas['Eficiencia_Operacional'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribuição da Eficiência Operacional')
plt.xlabel('Eficiência (%)')
plt.ylabel('Frequência')
plt.axvline(df_pandas['Eficiencia_Operacional'].mean(), color='red', linestyle='dashed', linewidth=2)
plt.show()
```

Calculamos a eficiência operacional média para cada terminal, seleciona os 10 terminais mais eficientes e os visualiza em um gráfico de barras horizontais. Ele agrupa os dados por terminal, calcula a média da eficiência operacional e ordena os resultados. Os 10 terminais com maior eficiência operacional são convertidos para Pandas e reordenados para exibição. O gráfico destaca os terminais mais eficientes e suas respectivas médias

```python
# Calcular a eficiência operacional média por terminal
df_terminal_eficiencia = df_cabotagem.groupBy('Terminal').agg(
    F.mean('Eficiencia_Operacional').alias('Media_Eficiencia_Operacional')
).orderBy('Media_Eficiencia_Operacional', ascending=False)

# Selecionar os top 10 terminais com maior eficiência operacional
top_terminais = df_terminal_eficiencia.limit(10).toPandas()

# Ordenar pelo valor da eficiência operacional
top_terminais = top_terminais.sort_values(by='Media_Eficiencia_Operacional', ascending=True)

plt.figure(figsize=(12,8))
plt.barh(top_terminais['Terminal'], top_terminais['Media_Eficiencia_Operacional'], color='green')
plt.xlabel('Eficiência Operacional Média (%)')
plt.ylabel('Terminal')
plt.title('Top 10 Terminais com Maior Eficiência Operacional')
plt.tight_layout()
plt.show()
```

Selecionamos os 10 terminais com menor eficiência operacional, ordenando-os pela eficiência de forma crescente. Após calcular a eficiência média por terminal, ele filtra os terminais com os valores mais baixos e os exibe em um gráfico de barras horizontais. O gráfico é colorido em vermelho e ajusta o limite do eixo X para focar em eficiencias operacionais entre 0% e 50%. Isso permite uma análise das unidades com pior desempenho operacional.

```python
# Selecionar os top 10 terminais com menor eficiência operacional
bottom_terminais = df_terminal_eficiencia.orderBy('Media_Eficiencia_Operacional', ascending=True).limit(10).toPandas()

# Ordenar pelo valor da eficiência operacional
bottom_terminais = bottom_terminais.sort_values(by='Media_Eficiencia_Operacional', ascending=True)

plt.figure(figsize=(12,8))
plt.barh(bottom_terminais['Terminal'], bottom_terminais['Media_Eficiencia_Operacional'], color='red')
plt.xlabel('Eficiência Operacional Média (%)')
plt.ylabel('Terminal')
plt.title('Top 10 Terminais com Menor Eficiência Operacional')

# Ajustando os limites do eixo X
plt.xlim(0, 50)

plt.tight_layout()
plt.show()
```

Calculamos diversos indicadores de desempenho das operações portuárias, como a Taxa de Ocupação do Berço e a Taxa de Utilização dos Recursos, além de medir o Tempo Médio de Atracação e Tempo Médio de Permanência por porto e terminal. Ele também avalia o Desempenho da Operação por tipo de operação, incluindo médias de Tempo de Operação, Tempo de Espera e Eficiência Operacional. Outros cálculos incluem o Tempo de Chegada até Início da Operação e o Tempo de Término da Desatracação, fornecendo uma análise abrangente da eficiência das operações portuárias.

```python
df_cabotagem = df_cabotagem.withColumn(
    'Taxa_Ocupacao_Berco',
    (unix_timestamp('Data Desatracação') - unix_timestamp('Data Atracação')) / (unix_timestamp('Data Desatracação') - unix_timestamp('Data Chegada')) * 100
)
df_atracacao_medio = df_cabotagem.groupBy('Porto Atracação').agg(
    F.mean((unix_timestamp('Data Desatracação') - unix_timestamp('Data Atracação')) / 3600).alias('Tempo_Medio_Atracacao')
).orderBy('Tempo_Medio_Atracacao', ascending=False)

df_permanencia_terminal = df_cabotagem.groupBy('Terminal').agg(
    F.mean((unix_timestamp('Data Desatracação') - unix_timestamp('Data Chegada')) / 3600).alias('Tempo_Medio_Permanencia')
).orderBy('Tempo_Medio_Permanencia', ascending=False)

df_permanencia_terminal = df_cabotagem.groupBy('Terminal').agg(
    F.mean((unix_timestamp('Data Desatracação') - unix_timestamp('Data Chegada')) / 3600).alias('Tempo_Medio_Permanencia')
).orderBy('Tempo_Medio_Permanencia', ascending=False)

df_cabotagem = df_cabotagem.withColumn(
    'Taxa_Utilizacao_Recursos',
    (unix_timestamp('Data Término Operação') - unix_timestamp('Data Início Operação')) / (unix_timestamp('Data Desatracação') - unix_timestamp('Data Chegada')) * 100
)

df_desempenho_operacao = df_cabotagem.groupBy('Tipo de Operação').agg(
    F.mean('Tempo_Operacao').alias('Media_Tempo_Operacao'),
    F.mean('Tempo_Espera').alias('Media_Tempo_Espera'),
    F.mean('Eficiencia_Operacional').alias('Media_Eficiencia_Operacional')
).orderBy('Media_Eficiencia_Operacional', ascending=False)

df_cabotagem = df_cabotagem.withColumn(
    'Tempo_Chegada_Inicio_Operacao',
    (unix_timestamp('Data Início Operação') - unix_timestamp('Data Chegada')) / 3600
)

df_cabotagem = df_cabotagem.withColumn(
    'Tempo_Termino_Desatracacao',
    (unix_timestamp('Data Desatracação') - unix_timestamp('Data Término Operação')) / 3600
)

```

## Previsão de Volume de TEUs com o Modelo Prophet

O código abaixo faz uma previsão do volume de TEUs (unidade de carga) utilizando o modelo **Prophet** e visualiza os dados históricos e as previsões para os próximos meses e anos.

### Descrição do processo:

1. **Preparação dos dados**: 
   - O DataFrame `df_cabotagem` é agrupado por mês, e a soma dos TEUs é calculada para cada mês (`AnoMes`).
   - O DataFrame resultante é então ajustado para que a coluna `ds` (data) seja formatada no formato `yyyy-MM-01` (adicionando o dia 1 a cada mês).

2. **Modelo Prophet**:
   - O modelo **Prophet** é ajustado aos dados históricos para prever a quantidade de TEUs para os próximos 15 meses.
   - O modelo é treinado considerando sazonalidades **anuais**, mas desconsidera as sazonais **semanais** e **diárias**.

3. **Previsões**:
   - O modelo gera previsões para os próximos meses e essas previsões são separadas em três períodos:
     - **Até setembro de 2024**: Previsões feitas com base nos dados reais.
     - **Outubro a dezembro de 2024**: Previsões para o final de 2024.
     - **2025**: Previsões para o ano de 2025.
   
4. **Intervalo de Confiança**:
   - As previsões incluem uma área sombreada representando a margem de erro, com base no intervalo de confiança (`yhat_lower` e `yhat_upper`).

5. **Visualização**:
   - **Dados reais**: Exibidos em uma linha contínua azul.
   - **Previsões até setembro de 2024**: Marcadas com pontos e uma linha preta.
   - **Previsões para outubro a dezembro de 2024**: Representadas por uma linha verde tracejada.
   - **Previsões para 2025**: Representadas por uma linha laranja tracejada.
   - A área de confiança é representada por uma faixa azul clara ao redor das previsões para o período de outubro de 2024 a dezembro de 2025.
   - O gráfico é ajustado com uma legenda, rótulos de eixo e um título destacado.

### Gráfico Resultante:

- **Eixo X**: Ano (com marcações mensais).
- **Eixo Y**: Volume de TEUs (quantidade movimentada).
- **Linhas**: Representam os dados reais e as previsões para os próximos períodos.
- **Área sombreada**: Intervalo de confiança para as previsões.

Este gráfico fornece uma visão clara da evolução do volume de TEUs, incluindo previsões confiáveis para os próximos meses e anos, o que pode ser útil para planejar operações logísticas e otimizar processos.


```python
# Suprimir mensagens de log desnecessárias do Prophet
logging.getLogger('cmdstanpy').setLevel(logging.WARNING)

# Converter o DataFrame Spark para Pandas para usar com o Prophet
df_monthly_teu = (
    df_cabotagem
    .withColumn("AnoMes", F.date_format(col("Data Atracação"), "yyyy-MM"))
    .groupBy("AnoMes")
    .agg(F.sum("TEU").alias("TEU"))
    .orderBy("AnoMes")
    .toPandas()
)

# Preparar os dados para o Prophet
df_monthly_teu['ds'] = pd.to_datetime(df_monthly_teu['AnoMes'] + "-01")  # Adicionando o dia 1 ao formato 'yyyy-MM'
df_monthly_teu['y'] = df_monthly_teu['TEU']
df_monthly_teu = df_monthly_teu[['ds', 'y']]

# Criar e treinar o modelo Prophet
model = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False)
model.fit(df_monthly_teu)

# Gerar previsões para 15 meses à frente
future_dates = model.make_future_dataframe(periods=15, freq='M')
forecast = model.predict(future_dates)

# Converter as previsões em um DataFrame Pandas para visualização
forecast_filtered = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].set_index('ds')

# Configuração do estilo do gráfico para um visual mais limpo e intuitivo
plt.style.use('ggplot')
plt.figure(figsize=(16, 9))

# Dados reais
plt.plot(
    df_monthly_teu['ds'],
    df_monthly_teu['y'],
    marker='o',
    linestyle='-',
    linewidth=2.0,
    color='dodgerblue',
    label='Dados Reais (2019-2024)'
)

# Previsões Prophet até setembro de 2024
forecast_to_sep = forecast_filtered.loc[:'2024-09-30']
plt.scatter(
    forecast_to_sep.index,
    forecast_to_sep['yhat'],
    color='black',
    label='Previsão Prophet (Até Set/2024)',
    zorder=3,
    s=50
)

# Previsões Prophet de outubro a dezembro de 2024
forecast_oct_to_dec = forecast_filtered.loc['2024-10-01':'2024-12-31']
plt.plot(
    forecast_oct_to_dec.index,
    forecast_oct_to_dec['yhat'],
    marker='o',
    linestyle='--',
    linewidth=2.0,
    color='green',
    label='Previsão Prophet (Out-Dez/2024)'
)

# Previsões Prophet para 2025
forecast_post_dec = forecast_filtered.loc['2025-01-01':]
plt.plot(
    forecast_post_dec.index,
    forecast_post_dec['yhat'],
    marker='o',
    linestyle='--',
    linewidth=2.0,
    color='orange',
    label='Previsão Prophet (2025)'
)

# Intervalo de confiança
forecast_with_ci = forecast_filtered.loc['2024-10-01':'2025-12-31']
plt.fill_between(
    forecast_with_ci.index,
    forecast_with_ci['yhat_lower'],
    forecast_with_ci['yhat_upper'],
    color='lightblue',
    alpha=0.3,
    label='Margem de Erro (Prophet, Out/2024 - Dez/2025)'
)

# Preenchendo o gap entre os dados reais e a previsão
last_actual_date = df_monthly_teu['ds'].max()
last_actual_value = df_monthly_teu[df_monthly_teu['ds'] == last_actual_date]['y'].values[0]
first_forecast_date = forecast_to_sep.index.min()
first_forecast_value = forecast_to_sep.loc[first_forecast_date]['yhat']

if last_actual_date + pd.DateOffset(months=1) == first_forecast_date:
    plt.plot(
        [last_actual_date, first_forecast_date],
        [last_actual_value, first_forecast_value],
        linestyle='-',
        color='dodgerblue',
        linewidth=2.0,
        zorder=1
    )

# Configurações do gráfico
plt.title('Evolução do Volume de TEUs com Previsão Prophet (2019-2025)', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Ano', fontsize=14, labelpad=10)
plt.ylabel('Volume de TEUs', fontsize=14, labelpad=10)
plt.legend(fontsize=12, loc='upper left', frameon=True)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

Através da análise sazonal e da aplicação do modelo Prophet, foi possível identificar padrões históricos de volume que refletem a dinâmica do setor ao longo dos anos.

As previsões geradas consideram a sazonalidade dos dados, mas é importante ressaltar que o modelo não leva em conta fatores externos como variações econômicas, mudanças políticas, climáticas ou outras variáveis exógenas que podem impactar significativamente as operações de cabotagem.

Embora os resultados ofereçam uma boa base para análise, reforça-se que as projeções são baseadas em tendências passadas e não devem ser interpretadas como garantias de comportamento futuro. Este projeto destaca o potencial das técnicas de machine learning e análise de séries temporais no estudo acadêmico de setores logísticos, ao mesmo tempo em que evidencia a importância de considerar variáveis externas para uma análise mais completa em estudos futuros.
